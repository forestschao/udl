```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi
```

```python
# Number of samples to generate for each distribution
N = 1000
dimensions = [25, 100, 500]

# Initialize plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, d in enumerate(dimensions):
  samples = np.random.randn(N, d)
  distances = np.linalg.norm(samples, axis=1)

   # Plot histogram
  axes[i].hist(distances, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')

  # Theoretical chi distribution
  x = np.linspace(chi.ppf(0.01, d), chi.ppf(0.99, d), 1000)
  pdf = chi.pdf(x, d)
  axes[i].plot(x, pdf, 'r-', lw=2, label=f'Chi(d={d})')

  axes[i].set_title(f'Dimension {d}')
  axes[i].set_xlabel('Distance')
  axes[i].set_ylabel('Density')
  axes[i].legend()

plt.tight_layout()
plt.show()
```
