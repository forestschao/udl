```python
import numpy as np
```

```python
def forward_pass(z, gamma, delta, epsilon=1e-5):
  f1 = z.mean()

  f2 = z - f1
  f3 = f2 * f2
  f4 = f3.mean()
  f5 = np.sqrt(f4 + epsilon)
  f6 = 1.0 / f5

  f7 = f2 * f6
  z_out = f7 * gamma + delta

  return z_out, f1, f2, f3, f4, f5, f6, f7
```

```python
def backward_pass(dz, z, f1, f2, f3, f4, f5, f6, f7, gamma, delta):
  N = z.shape

  dsigma = np.sum(dz)
  dgamma = np.sum(f7 * dz)

  df7i = gamma * dz
  df6  = np.sum(f2 * df7i)
  df5  = - 1.0 / np.square(f5) * df6
  df4  = 1.0 / (2.0 * f5) * df5
  df3i = df4 / N
  df2i = 2 * f2 * df3i + f6 * df7i
  df1  = -np.sum(df2i)
  dzi  = df2i + 1 / N * df1

  return dzi, dsigma, dgamma
```
